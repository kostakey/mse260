```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. Define your data (Example values for semiconductors)
materials = ['MgO', 'CaO', 'SrO', 'BaO']
lattice_params = [4.19, 4.81, 5.17, 5.58]  # in Angstroms
band_gaps = [4.43, 3.63, 3.27, 2.09]       # in eV

# 2. Calculate the Line of Best Fit (Linear Regression)
# polyfit(x, y, 1) returns [slope, intercept] for a 1st degree polynomial
slope, intercept = np.polyfit(lattice_params, band_gaps, 1)

# Generate values for the line
x_fit = np.linspace(min(lattice_params), max(lattice_params), 100)
y_fit = slope * x_fit + intercept

# 2. Create the plot
plt.figure(figsize=(10, 6))

# Use scatter for distinct points, or plot() for lines
plt.scatter(lattice_params, band_gaps, color='royalblue', s=80, edgecolors='k', label='Rock Salt Crystals')

# Plot the Line of Best Fit
plt.plot(x_fit, y_fit, color='crimson', linestyle='--', linewidth=2, 
         label=f'Best Fit: $E_g = {slope:.2f}a + {intercept:.2f}$')

# 3. Add labels (using LaTeX for mathematical symbols)
plt.xlabel(r'Lattice Parameter $a$ ($\AA$)', fontsize=14)
plt.ylabel(r'Band Gap $E_g$ ($eV$)', fontsize=14)
plt.title(r'Band Gap vs. Lattice Parameter', fontsize=16)

# 4. Optional: Annotate points with the material names
for i, name in enumerate(materials):
    plt.annotate(name, (lattice_params[i], band_gaps[i]), 
                 xytext=(5, 5), textcoords='offset points', fontsize=10)

# 5. Styling and Grid
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend()

# 6. Save or Display
# plt.savefig('band_gap_plot.png', dpi=300)
plt.show()
```

```python
# 1. Define your data (Example values for semiconductors)
materials = ['MgO', 'CaO', 'SrO', 'BaO']
melting_temp = [3125, 2886, 2804, 2196]  # in Kelvin
band_gaps = [4.43, 3.63, 3.27, 2.09]       # in eV

# 2. Calculate the Line of Best Fit (Linear Regression)
# polyfit(x, y, 1) returns [slope, intercept] for a 1st degree polynomial
slope, intercept = np.polyfit(melting_temp, band_gaps, 1)

# Generate values for the line
x_fit = np.linspace(min(melting_temp), max(melting_temp), 100)
y_fit = slope * x_fit + intercept

# print(y_fit)

# 2. Create the plot
plt.figure(figsize=(10, 6))

# Use scatter for distinct points, or plot() for lines
plt.scatter(melting_temp, band_gaps, color='royalblue', s=80, edgecolors='k', label='Rock Salt Crystals')

# Plot the Line of Best Fit
plt.plot(x_fit, y_fit, color='crimson', linestyle='--', linewidth=2, 
         label=f'Best Fit: $E_g = {slope:.4f}T_m {intercept:.2f}$')

# 3. Add labels (using LaTeX for mathematical symbols)
plt.xlabel(r'Melting Temperature $T_m$ ($K$)', fontsize=14)
plt.ylabel(r'Band Gap $E_g$ ($eV$)', fontsize=14)
plt.title(r'Band Gap vs. Melting Temperature', fontsize=16)

# 4. Optional: Annotate points with the material names
for i, name in enumerate(materials):
    plt.annotate(name, (melting_temp[i], band_gaps[i]), 
                 xytext=(5, 5), textcoords='offset points', fontsize=10)

# 5. Styling and Grid
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend()

# 6. Save or Display
# plt.savefig('band_gap_plot.png', dpi=300)
plt.show()
```
